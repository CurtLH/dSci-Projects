```python
import pandas as pd
import geotext
import numpy as np
```

```python
# load data
df = pd.read_excel('20180718_parsed_alabama_locations.xlsx')
```

```python
# inspect data
df.head()
```

```python
# extract cities from location field
def get_city(row):
    try:
        place = geotext.GeoText(row['Location'])
        cities = set(i.title() for i in place.cities)
        return cities

    except:
        return np.nan
    
df['Location_City'] = df.apply(lambda row: get_city(row), axis=1) 
```
